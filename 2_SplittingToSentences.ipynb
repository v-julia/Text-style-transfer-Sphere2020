## Разбиение текстов на предложения

Стихотворения, скачанные в ноутбуке LoadingPoems, были вручную проверены. Стихотворения с матными словами и нераспознанным текстом были вручную удалены. В  данном ноутбуке производится дальнейшая обработка текстов и разбиения их на предложения.

```python
import nltk.data
import random
from tqdm import tqdm
```

```python
from nltk.tokenize import word_tokenize
```

```python
DATA_PATH = './data/'
```

Загрузка стихотворений

```python
def load_poems(file_path):
    poems = []
    with open(file_path, encoding='UTF-8') as f:
        poem_text = []
        for line in f:
            line = line.replace('\n', ' ')
            if line == '------------- ':
                poems.append(poem_text)
                poem_text = []
            else:
                poem_text.append(line)
    f.close()
    return poems
```

```python
import re

eng = re.compile(r'[A-Za-z]')
punct = re.compile(r'[?!.]')
```

В некоторых стихотворениях встречаются:

- иностранные слова (очень маленькая частота)
- слова со смесью русских и английских букв, похожих на русские (некорректное распознавание текста)
- знаки пунктуации в словах, которые плохо распознались ("вс?")

Поэтому все слова в предложении проверяем на эти условия.

```python
def review_sent(sentences):
    sentences_rev = []
    for s in sentences:
        words = [word for word in word_tokenize(s) if word.isalnum()]

        flag = 1
        for word in words:
            if eng.search(word) or punct.search(word):
                flag = 0
                print(word)
                #print(s)
        if flag:
            sentences_rev.append(s)
    return sentences_rev
```

Длинные предложения откладываем отдельно, чтобы рещить, как их потом делить.

```python
def create_sentences(poems_sentences):
    texts = []
    long_sent = []
    for poem in poems_sentences:
        cur_text = ''
        cur_len = 0
        for s in poem:
            sent_len = len([word for word in word_tokenize(s) if word.isalnum()])
            if sent_len > 80:
                long_sent.append(s)
                print(sent_len)
                continue
            if cur_len + sent_len > 35:
                texts.append(cur_text)
                cur_text = s
                cur_len = sent_len
            else:
                cur_text = cur_text + ' ' + s
                cur_len += sent_len
        texts.append(cur_text)
    return texts, long_sent
```

```python
def save_sentences(file_path, sentences):
    with open(file_path, 'w', encoding='UTF-8') as f:
        for s in sentences:
            f.write(s)
            f.write('\n')
    f.close()
```
